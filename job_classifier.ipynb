## Part I: Labeling

```python
# Drop rows with a missing value in org or role_original

df = df.dropna(subset=['org', 'role_original'])
```

```python
df.shape
```

```python
# Convert role_original to lowercase

df.loc[:, 'role_original'] = df['role_original'].str.lower()
```

```python
df.head(20)
```

```python
# List of Departments
departments = ['Investment Banking', 'Asset/Wealth Management', 'Sales & Trading', 'Retail Banking', 'Corporate Banking', 'Risk Management', 'Finance', 'Marketing', 'Insurance', 'IT', 'Other']
```

```python
# If role_original contains agent or sales, then set department to Retail Banking

df.loc[df['role_original'].str.contains('agent|sales'), 'department'] = 'Retail Banking'
df.head(10)
```

```python
# If role_original contains financial advisor, then set department to Financial Advisory

df.loc[df['role_original'].str.contains('financial advisor'), 'department'] = 'Financial Advisory'
```

```python
df.head(10)
```

```python
# If role_original contains customer service, then set department to Customer Support

df.loc[df['role_original'].str.contains('customer service'), 'department'] = 'Customer Support'
df.head()
```

```python
# If role_original contains accountant, then set department to Finance
df.loc[df['role_original'].str.contains('accountant'), 'department'] = 'Finance'

# If role_original contains software engineer, then set department to IT
df.loc[df['role_original'].str.contains('software engineer|technology|develop'), 'department'] = 'IT'

# If role_original contains Teller, Banker, Mortgage, Representative, Branch, or Banking, then set department to Retail Banking
df.loc[df['role_original'].str.contains('teller|banker|mortgage|branch|banking'), 'department'] = 'Retail Banking'

# If role_original contains claim, then set department to Claims Management
df.loc[df['role_original'].str.contains('claim'), 'department'] = 'Claims Management'

# If role_original contains investment, then set department to Corporate & Investment Banking
df.loc[df['role_original'].str.contains('investment'), 'department'] = 'Corporate & Investment Banking'

# If role_original contains insurance, then set department to Insurance
df.loc[df['role_original'].str.contains('insurance'), 'department'] = 'Insurance'

# If role_original contains wealth, then set department to Wealth Management
df.loc[df['role_original'].str.contains('wealth'), 'department'] = 'Wealth Management'

# If role_original contains consultant, then set department to Consulting
df.loc[df['role_original'].str.contains('consultant'), 'department'] = 'Consulting'
```

```python
# prompt: sort based on role_original

df_sorted = df.sort_values('role_original')
df_sorted
```

```python
# Exclude rows with role_original with special characters

import re
pattern = r"[^a-zA-Z0-9\s]"
df = df[~df['role_original'].str.contains(pattern)]
df
```

```python
words_more_context = ['manager', 'director', 'associate manager', 'associate director', 'intern']
CIB = ['J.P.Morgan', 'UBS', 'Deutche Bank', 'Credit Suisse', 'Goldman Sachs']
RB = ['HSBC', 'Citi', 'RBC', 'Bank of America']
```

```python
# If role_original contains a word in words_more_context AND the company name is within CIB,
# then set department to Corporate & Investment Banking.
# If the company name is within RB, then set department to Retail Banking.

for word in words_more_context:
  df.loc[(df['role_original'] == (word)) & (df['org'].isin(CIB)), 'department'] = 'Corporate & Investment Banking'
  df.loc[(df['role_original'] == (word)) & (df['org'].isin(RB)), 'department'] = 'Retail Banking'
```

```python
# Number of departments not null

print(df['department'].notnull().sum())
```

```python
# Export df as a csv file for manual data labeling
%cd /root
df.to_csv('labeled_org_role_100K_2.csv', index=False)
files.download('labeled_org_role_100K_2.csv')
```
